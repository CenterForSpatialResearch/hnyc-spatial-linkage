### Understand Block Number

```python
def uniqueness(df, level, col):
    df_check = df[[level, col]]
    df2 = df_check.groupby(level).apply(lambda x: x[col].unique())
    d = df2.to_dict()
    for key1 in d:
        for key2 in d:
            if key1 != key2:
                check = any(item in d[key1] for item in d[key2])
                if check is True:
                    print(key1, key2)
                    raise Exception(str(col) + " numbers are not unique")
    return True
```

```python
uniqueness(filled_1850, "WARD_NUM","block_num")
```

### Let's Try Adding in Block Numbers for a Simple Linear Model to Predict House Number and Seeing how well that works

```python
#get dwellings level dataset
dwellings_1850 = filled_1850.groupby(["WARD_NUM", "CENSUS_DWELLING_NUM"], as_index = False).first()
dwellings_1880 = census_1880.drop_duplicates(subset = ["CENSUS_ADDRESS"]).copy()
```

```python
#updated implementation of this with groupby(WARD_NUM) should be used with future code
sequences.col_for_seq(dwellings_1850, "CD_X", "CD_Y")
sequences.col_for_seq(dwellings_1880, "POINT_X", "POINT_Y")
dwellings_1850 = sequences.get_dist_seq(dwellings_1850, 0.15)
dwellings_1880 = sequences.get_dist_seq(dwellings_1880, 0.15)
dwellings_1850.groupby("sequence_id").apply(sequences.sequence_order)
dwellings_1880.groupby("sequence_id").apply(sequences.sequence_order)
census_1880_model = dataprocessing.dwellings_to_all(census_1880, dwellings_1880, ["CENSUS_MATCH_ADDR", "sequence_id"], ["CENSUS_MATCH_ADDR"])
census_1850_model = dataprocessing.dwellings_to_all(filled_1850, dwellings_1850, ["WARD_NUM", "CENSUS_DWELLING_NUM","sequence_id"], ["WARD_NUM", "CENSUS_DWELLING_NUM"])
dataprocessing.create_street_house(census_1880_model, "CENSUS_ADDRESS")
dataprocessing.create_street_house(census_1850_model, "CD_ADDRESS")
```

```python
cols = ["street_name", "sequence_id", "WARD_NUM", "CENSUS_SERIALB", "block_num"]
train_X, train_y, test_1880_X, test_1880_y, test_1850_X, test_1850_y = archive.create_train_test_data(census_1880_model, census_1850_model, cols)
```

```python
targetencoder = TargetEncoder(cols = cols)
train_X = targetencoder.fit_transform(train_X, train_y)
test_1880_X = targetencoder.fit_transform(test_1880_X, test_1880_y)
test_1850_X = targetencoder.fit_transform(test_1850_X, test_1850_y)
```

```python
model_lin = LinearRegression()
model_lin.fit(train_X, train_y)
print("Training score:", model_lin.score(train_X, train_y))
print("Test 1850 score:", model_lin.score(test_1850_X, test_1850_y))
print("Test 1880 score:", model_lin.score(test_1880_X, test_1880_y))
```

```python
archive.graph_coefs(targetencoder.get_feature_names(), model_lin.coef_, "Regression Model with Enumeration Based on Geography Order")
```
