```python
before = pd.read_csv('../../census_enum_seq_df_05.csv')
after = pd.read_csv('../../census_enum_seq_df_05_debugged.csv')

before = before[['block_num', 'CD_ADDRESS', 'cd_X', 'cd_Y', 'dwelling_id',
                     'sequence_id', 'enum_dist_id', 'enum_seq', 'dist']].dropna(subset=['cd_X', 'cd_Y'])\
                                    .drop_duplicates()
after = after[['block_num', 'CD_ADDRESS', 'cd_X', 'cd_Y', 'dwelling_id',
                     'sequence_id', 'enum_dist_id', 'enum_seq', 'dist']].dropna(subset=['cd_X', 'cd_Y'])\
                                    .drop_duplicates()
# before['enum_dist_id'].replace('Ward_Num', -1, inplace=True)
# before['enum_dist_id'] = before['enum_dist_id'].astype('float64')
```
