### Let's Look At Some Wards in Both 1850 and 1880

```python
census_1850_filled = pd.read_csv("data/Census_1850_Matches.csv")
```

```python
for ward in census_1850["CENSUS_WARD_NUM"].unique():
    fig, ax = plt.subplots(1,2)
    ward1 = census_1880[census_1880["WARD_NUM"] == ward]
    ward2 = census_1850_filled[census_1850_filled["CENSUS_WARD_NUM"] == ward]
    ward1.plot(x = "CENSUS_X", y = "CENSUS_Y", ax = ax[0])
    ward2.plot(x = "CD_X", y = "CD_Y", ax = ax[1])
    fig.tight_layout()
    if ward == 10:
        break
```

From this rough comparison, it doesn't seem that we can be confident about the similarities between 1850 and 1880, but how much these differences will matter is something that I'm pretty unsure about too. One issues is that of course we have much fewer addresses for 1880 only about 38% of dwellings are matched which makes comparison difficult.
